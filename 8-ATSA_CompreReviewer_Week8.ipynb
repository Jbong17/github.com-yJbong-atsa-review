## **PART 1 — Concept Map: How Everything Connects**

> Every topic in this course builds on or connects to others.
> The exam tests whether you see the **whole structure**, not just isolated facts.

---

### **1.1 The Foundation Chain**

$$
\text{Stationarity} \;\longrightarrow\; \text{ACF/PACF} \;\longrightarrow\; \text{Wold Decomposition} \;\longrightarrow\; \text{ARMA}
$$

- **Stationarity** (Week 1) makes ACF/PACF well-defined
- **ACF/PACF** (Week 1) summarize linear dependence and guide identification
- **Wold** (Week 1) guarantees that ARMA models are theoretically justified
- **ARMA** (Weeks 1–2) is the practical realization of Wold's infinite MA

---

### **1.2 Time Domain ↔ Frequency Domain Duality**

$$
\gamma(h) \;\xleftrightarrow{\text{Fourier}}\; f(\omega)
$$

| Time domain | Frequency domain |
|---|---|
| ACF decays slowly | Spectral peak at $\omega = 0$ |
| ACF cuts off at lag $q$ | Smooth spectrum (MA structure) |
| Differencing removes trend | High-pass filter suppresses $\omega = 0$ |
| Seasonal differencing | Zeros at seasonal harmonics |
| ARMA spectral density | $f(\omega) = \sigma^2 |\theta(e^{-i\omega})|^2 / |\phi(e^{-i\omega})|^2$ |

---

### **1.3 Univariate → Multivariate Extension**

| Univariate (Weeks 1–4) | Multivariate (Week 5) |
|---|---|
| AR($p$) | VAR($p$) |
| ACF/PACF | Cross-correlation, Granger causality |
| $\psi$-weights | IRF matrices $\boldsymbol{\Psi}_j$ |
| Unit root → differencing | Unit root → cointegration / VECM |
| ARMA forecasting | VAR forecasting + FEVD |

---

### **1.4 State Space as the Unifying Framework**

| Model | State space form |
|---|---|
| AR(1) | $\Phi = \phi$, $A = 1$, $Q = \sigma^2$, $R = 0$ |
| ARMA(1,1) | 2D state vector |
| Local level | $\Phi = 1$ (random walk state), $R > 0$ |
| Structural (trend + season) | Multi-dimensional state |
| VAR | Multi-dimensional state + observation |

The **Kalman filter** provides optimal estimation for all of these.

---

### **1.5 Linear vs Nonlinear Methods**

| Linear (Weeks 1–6) | Nonlinear (Week 7) |
|---|---|
| ARMA, VAR, state space | Simplex projection, S-maps |
| Granger causality (predictive, linear) | CCM (dynamical, nonlinear) |
| Spectral density | Attractor reconstruction |
| Assumes stochastic noise | Exploits deterministic structure |
| Wold decomposition | Takens' theorem |

---

### **1.6 The Forecasting Hierarchy**

$$
\text{Naive} \;\longrightarrow\; \text{ARIMA} \;\longrightarrow\; \text{SARIMA} \;\longrightarrow\; \text{VAR/VECM} \;\longrightarrow\; \text{State Space} \;\longrightarrow\; \text{Nonlinear/ML}
$$

Each level adds complexity. The exam tests whether you know **when each is appropriate**
and **what breaks when assumptions fail**.


## **PART 2 — Formula Reference Sheet**

> Print this section for quick review. All key results across Weeks 1–7.

---

### **Stationarity & Dependence (Week 1)**

| Quantity | Formula |
|---|---|
| Weak stationarity | $\mathbb{E}[X_t] = \mu$, $\mathrm{Var}(X_t) = \sigma^2$, $\mathrm{Cov}(X_t, X_{t+k}) = \gamma(k)$ |
| ACF | $\rho(k) = \gamma(k) / \gamma(0)$ |
| AR(1) variance | $\gamma(0) = \sigma^2 / (1 - \phi^2)$ |
| AR(1) ACF | $\rho(k) = \phi^k$ |
| MA(1) ACF | $\rho(1) = \theta/(1+\theta^2)$, $\rho(k) = 0$ for $k \geq 2$ |
| AR(1) stationarity | $|\phi| < 1$ |
| MA(1) invertibility | $|\theta| < 1$ |

---

### **Box–Jenkins & SARIMA (Week 2)**

| Concept | Formula / Rule |
|---|---|
| ARIMA($p,d,q$) | $\phi(B)\nabla^d X_t = \theta(B)\varepsilon_t$ |
| SARIMA | $\phi(B)\Phi(B^s)\nabla^d\nabla_s^D X_t = \theta(B)\Theta(B^s)\varepsilon_t$ |
| AIC | $-2\ln\hat{L} + 2k$ |
| BIC | $-2\ln\hat{L} + k\ln n$ |
| Model adequacy rule | Diagnostics first, then AIC/BIC to rank adequate models |

---

### **Forecasting (Week 3)**

| Quantity | Formula |
|---|---|
| AR(1) $h$-step forecast | $\hat{x}_{t+h|t} = \phi^h X_t$ |
| AR(1) error variance | $\mathrm{Var}(e_{t+h}) = \sigma^2(1-\phi^{2h})/(1-\phi^2)$ |
| ARMA(1,1) $\psi$-weights | $\psi_0=1$, $\psi_j = \phi^{j-1}(\phi+\theta)$ for $j \geq 1$ |
| General error variance | $\mathrm{Var}(e_{t+h}) = \sigma^2[1 + \sum_{j=1}^{h-1}\psi_j^2]$ |
| Prediction interval | $\hat{x}_{t+h|t} \pm z_{1-\alpha/2}\sqrt{\mathrm{Var}(e_{t+h})}$ |

---

### **Spectral Analysis (Week 4)**

| Quantity | Formula |
|---|---|
| Spectral density | $f(\omega) = \sum_{h=-\infty}^{\infty}\gamma(h)e^{-2\pi i\omega h}$ |
| Inverse (Wiener–Khinchin) | $\gamma(h) = \int_{-1/2}^{1/2}f(\omega)e^{2\pi i\omega h}d\omega$ |
| Total variance | $\gamma(0) = \int f(\omega)d\omega$ |
| AR(1) spectrum | $f(\omega) = \sigma^2/(1 - 2\phi\cos 2\pi\omega + \phi^2)$ |
| MA(1) spectrum | $f(\omega) = \sigma^2(1 + 2\theta\cos 2\pi\omega + \theta^2)$ |
| Periodogram | $I(\omega_j) = |d(\omega_j)|^2/n$ (inconsistent!) |
| Filter effect | $f_Y(\omega) = |A(\omega)|^2 f_X(\omega)$ |
| First diff transfer | $|1-e^{-2\pi i\omega}|^2 = 2(1-\cos 2\pi\omega)$ |

---

### **Multivariate (Week 5)**

| Concept | Formula / Rule |
|---|---|
| VAR(1) | $\mathbf{X}_t = \mathbf{A}\mathbf{X}_{t-1} + \boldsymbol{\varepsilon}_t$ |
| VAR stability | All eigenvalues of $\mathbf{A}$ inside unit circle |
| IRF | $\boldsymbol{\Psi}_j = \mathbf{A}^j$ (for VAR(1)) |
| Granger causality | F-test: $H_0$: cross-lag coefficients = 0 |
| Cointegration | $Y_t - \beta X_t \sim I(0)$ for $I(1)$ series |
| VECM | $\Delta\mathbf{X}_t = \boldsymbol{\alpha}\boldsymbol{\beta}'\mathbf{X}_{t-1} + \sum\boldsymbol{\Gamma}_j\Delta\mathbf{X}_{t-j} + \boldsymbol{\varepsilon}_t$ |
| GARCH(1,1) | $\sigma_t^2 = \alpha_0 + \alpha_1 X_{t-1}^2 + \beta_1\sigma_{t-1}^2$; stationarity: $\alpha_1+\beta_1 < 1$ |

---

### **State Space & Kalman (Week 6)**

| Step | Formula |
|---|---|
| State equation | $\mathbf{x}_t = \boldsymbol{\Phi}\mathbf{x}_{t-1} + \mathbf{w}_t$ |
| Observation equation | $y_t = \mathbf{A}'\mathbf{x}_t + v_t$ |
| KF prediction | $\mathbf{x}_{t|t-1} = \boldsymbol{\Phi}\mathbf{x}_{t-1|t-1}$; $\mathbf{P}_{t|t-1} = \boldsymbol{\Phi}\mathbf{P}_{t-1|t-1}\boldsymbol{\Phi}' + \mathbf{Q}$ |
| Innovation | $\nu_t = y_t - \mathbf{A}'\mathbf{x}_{t|t-1}$; $S_t = \mathbf{A}'\mathbf{P}_{t|t-1}\mathbf{A} + R$ |
| Kalman gain | $\mathbf{K}_t = \mathbf{P}_{t|t-1}\mathbf{A} S_t^{-1}$ |
| KF update | $\mathbf{x}_{t|t} = \mathbf{x}_{t|t-1} + \mathbf{K}_t\nu_t$; $\mathbf{P}_{t|t} = (\mathbf{I}-\mathbf{K}_t\mathbf{A}')\mathbf{P}_{t|t-1}$ |
| Exact log-likelihood | $\ln L = -\frac{1}{2}\sum[\ln 2\pi + \ln S_t + \nu_t^2/S_t]$ |
| Covariance ordering | $P_{\text{smooth}} \leq P_{\text{filter}} \leq P_{\text{predict}}$ |

---

### **Nonlinear & CCM (Week 7)**

| Concept | Key Idea |
|---|---|
| Takens' theorem | Embedding $(x_t, x_{t-\tau}, \ldots, x_{t-(E-1)\tau})$ reconstructs attractor |
| Simplex projection | Nearest-neighbor forecast on shadow attractor; optimal $E$ = attractor dimension |
| S-maps | $\theta=0$: linear; $\theta>0$: nonlinear; test by comparing forecast skill |
| CCM direction | To test "$X$ causes $Y$": cross-map from $M_Y$ to $X$ |
| CCM convergence | True causality → cross-map $\rho$ increases with library size $L$ |
| Granger vs CCM | Granger = linear/stochastic; CCM = nonlinear/deterministic; use both |


```python
# ============================================================
# SETUP
# ============================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy.stats import norm

np.random.seed(123)
print("Setup complete.")
```

## **PART 3 — End-to-End Diagnostic Pipeline**

> This section demonstrates the **complete workflow** from raw data to final model defense,
> integrating techniques from all weeks. This is what an exam answer should look like
> when given a dataset and asked to "analyze this time series."

---

### **The Pipeline**

1. **Inspect** (Week 1): plot, check stationarity, compute ACF/PACF
2. **Transform** (Week 2): log-transform, difference if needed
3. **Identify** (Week 2): propose candidates from ACF/PACF patterns
4. **Spectral check** (Week 4): periodogram to confirm periodicities
5. **Fit** (Week 2): estimate ARIMA/SARIMA candidates
6. **Diagnose** (Week 2): residual whiteness, normality, Ljung–Box
7. **Forecast** (Week 3): predict with PIs, backtest against baselines
8. **Defend** (Week 2): articulate why this model, not another


```python
# ============================================================
# PART 3: End-to-End Pipeline on Simulated PM2.5
# ============================================================

np.random.seed(123)
n = 240
period = 12
t = np.arange(n)
idx = pd.date_range("2005-01-01", periods=n, freq="MS")

baseline = np.log(25)
trend_log = -0.0015 * t
season_log = 0.25 * np.cos(2*np.pi*t/period) - 0.15 * np.sin(2*np.pi*t/period)
spike = np.zeros(n)
spike_months = np.random.choice(np.arange(12, n-12), size=10, replace=False)
spike[spike_months] = np.random.uniform(0.4, 0.9, size=len(spike_months))
eps = np.random.normal(0, 0.22, size=n)
log_pm25 = baseline + trend_log + season_log + spike + eps
y = pd.Series(np.exp(log_pm25), index=idx, name="PM25")

# ===== STEP 1: INSPECT =====
print("=" * 60)
print("STEP 1: INSPECT")
print("=" * 60)

fig, axes = plt.subplots(2, 2, figsize=(14, 7))

axes[0, 0].plot(y, linewidth=1, color="#2c3e50")
axes[0, 0].set_title("Raw PM2.5", fontsize=11)
axes[0, 0].grid(True, alpha=0.3)

y_log = np.log(y)
axes[0, 1].plot(y_log, linewidth=1, color="#2980b9")
axes[0, 1].set_title("log(PM2.5) — variance stabilized", fontsize=11)
axes[0, 1].grid(True, alpha=0.3)

plot_acf(y_log.dropna(), lags=48, ax=axes[1, 0])
axes[1, 0].set_title("ACF of log(PM2.5)", fontsize=10)
plot_pacf(y_log.dropna(), lags=48, method="ywm", ax=axes[1, 1])
axes[1, 1].set_title("PACF of log(PM2.5)", fontsize=10)
plt.tight_layout()
plt.show()

stat, pval, _, _, _, _ = adfuller(y_log.dropna(), autolag="AIC")
print(f"ADF on log(PM2.5): stat={stat:.4f}, p={pval:.4f}")
print(f"Result: {'Stationary' if pval < 0.05 else 'Nonstationary -> differencing needed'}")

# ===== STEP 2: TRANSFORM + DIFFERENCE =====
print(f"\n{'=' * 60}")
print("STEP 2: TRANSFORM + DIFFERENCE")
print("=" * 60)

d1_d12 = y_log.diff(12).diff(1).dropna()
stat2, pval2, _, _, _, _ = adfuller(d1_d12, autolag="AIC")
print(f"ADF on nabla*nabla_12 log(PM2.5): stat={stat2:.4f}, p={pval2:.4f}")
print(f"Result: {'Stationary' if pval2 < 0.05 else 'Still nonstationary'}")

# ===== STEP 3: IDENTIFY =====
print(f"\n{'=' * 60}")
print("STEP 3: IDENTIFY CANDIDATES")
print("=" * 60)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(d1_d12, lags=36, ax=axes[0])
axes[0].set_title("ACF: $\\nabla\\nabla_{12}$ log(PM2.5)", fontsize=10)
plot_pacf(d1_d12, lags=36, method="ywm", ax=axes[1])
axes[1].set_title("PACF: $\\nabla\\nabla_{12}$ log(PM2.5)", fontsize=10)
plt.tight_layout()
plt.show()

print("Reading ACF/PACF of differenced series:")
print("  Short lags: ACF spike at lag 1 -> MA(1) non-seasonal component")
print("  Lag 12: ACF spike -> SMA(1) seasonal component")
print("  Candidates: SARIMA(0,1,1)(0,1,1)_12, SARIMA(1,1,1)(0,1,1)_12")

# ===== STEP 4: SPECTRAL CHECK =====
print(f"\n{'=' * 60}")
print("STEP 4: SPECTRAL CHECK")
print("=" * 60)

x_spec = y_log.dropna().values - np.mean(y_log.dropna().values)
n_spec = len(x_spec)
freqs = np.fft.rfftfreq(n_spec, d=1.0)
Pxx = np.abs(np.fft.rfft(x_spec))**2 / n_spec

plt.figure(figsize=(12, 3))
plt.plot(freqs[1:], Pxx[1:], linewidth=0.8, color="#2c3e50")
plt.axvline(1/12, color="red", linestyle="--", alpha=0.7, label="Annual ($\\omega=1/12$)")
plt.axvline(2/12, color="orange", linestyle="--", alpha=0.5, label="Semi-annual")
plt.title("Periodogram of log(PM2.5) — confirms seasonal peaks", fontsize=12)
plt.xlabel("Frequency"); plt.ylabel("Power")
plt.legend(fontsize=9); plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ===== STEP 5: FIT =====
print(f"\n{'=' * 60}")
print("STEP 5: FIT SARIMA CANDIDATES")
print("=" * 60)

candidates = {
    "SARIMA(0,1,1)(0,1,1)12": {"order": (0,1,1), "seasonal_order": (0,1,1,12)},
    "SARIMA(1,1,1)(0,1,1)12": {"order": (1,1,1), "seasonal_order": (0,1,1,12)},
    "SARIMA(1,1,0)(1,1,0)12": {"order": (1,1,0), "seasonal_order": (1,1,0,12)},
}

fits = {}
for name, spec in candidates.items():
    try:
        res = SARIMAX(y_log, order=spec["order"], seasonal_order=spec["seasonal_order"],
                      enforce_stationarity=True, enforce_invertibility=True).fit(disp=False)
        fits[name] = res
        print(f"  {name}: AIC={res.aic:.2f}, BIC={res.bic:.2f}")
    except Exception as e:
        print(f"  {name}: FAILED ({e})")

# ===== STEP 6: DIAGNOSE =====
print(f"\n{'=' * 60}")
print("STEP 6: RESIDUAL DIAGNOSTICS")
print("=" * 60)

best_name = min(fits, key=lambda k: fits[k].aic)
best_res = fits[best_name]
r = best_res.resid.dropna()

fig, axes = plt.subplots(1, 3, figsize=(15, 3.5))

axes[0].plot(r, linewidth=0.8, color="#2c3e50")
axes[0].axhline(0, color="gray", linestyle="--", alpha=0.5)
axes[0].set_title(f"{best_name} — Residuals", fontsize=10)
axes[0].grid(True, alpha=0.3)

a_r = acf(r, nlags=36, fft=True)
ci = 1.96 / np.sqrt(len(r))
axes[1].stem(range(len(a_r)), a_r, basefmt=" ", markerfmt=".", linefmt="-")
axes[1].axhline(ci, color="red", linestyle="--", alpha=0.5)
axes[1].axhline(-ci, color="red", linestyle="--", alpha=0.5)
axes[1].set_title("Residual ACF", fontsize=10)
axes[1].grid(True, alpha=0.3)

axes[2].hist(r, bins=25, density=True, alpha=0.7, color="#3498db", edgecolor="white")
xg = np.linspace(r.min(), r.max(), 200)
axes[2].plot(xg, norm.pdf(xg, r.mean(), r.std()), "r-", linewidth=2)
axes[2].set_title("Residual Distribution", fontsize=10)
axes[2].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

lb = acorr_ljungbox(r, lags=[12, 24, 36], return_df=True)
print(f"\nLjung-Box for {best_name}:")
print(lb.to_string())
passes = not (lb["lb_pvalue"] < 0.05).any()
print(f"Whiteness: {'PASS' if passes else 'FAIL'}")

# ===== STEP 7: FORECAST =====
print(f"\n{'=' * 60}")
print("STEP 7: FORECAST (12-month ahead)")
print("=" * 60)

fc = best_res.get_forecast(steps=12)
fc_mean = fc.predicted_mean
fc_ci = fc.conf_int(alpha=0.05)

plt.figure(figsize=(14, 4))
plt.plot(y_log.index, y_log.values, linewidth=1, color="#2c3e50", label="Observed")
plt.plot(fc_mean.index, fc_mean.values, linewidth=2, color="#e74c3c", label="Forecast")
plt.fill_between(fc_ci.index, fc_ci.iloc[:, 0], fc_ci.iloc[:, 1],
                  alpha=0.2, color="#e74c3c", label="95% PI")
plt.title(f"{best_name} — 12-Month Forecast", fontsize=12)
plt.legend(fontsize=9); plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ===== STEP 8: DEFEND =====
print(f"\n{'=' * 60}")
print("STEP 8: MODEL DEFENSE")
print("=" * 60)
print(f"""
Final model: {best_name}

Defense:
  1. Log-transformed to stabilize multiplicative variance.
  2. Inspected ACF/PACF: slow decay + seasonal spikes -> nonstationarity + seasonality.
  3. Applied regular + seasonal differencing (d=1, D=1, s=12).
  4. Periodogram confirmed annual peak at omega=1/12.
  5. Proposed 3 SARIMA candidates from ACF/PACF of differenced series.
  6. Selected {best_name} (lowest AIC={best_res.aic:.2f}).
  7. Residuals pass Ljung-Box at lags 12/24/36 -> adequate.
  8. Model is parsimonious and diagnostically adequate.
""")
```

### **Block C: Spectral and Frequency Domain (Questions 8–9)**

**Q8.** *State the Wiener–Khinchin theorem. Why is the raw periodogram an inconsistent estimator?*

**Model answer keywords:** $f(\omega) = \sum\gamma(h)e^{-2\pi i\omega h}$, inverse recovers $\gamma(h)$; periodogram: unbiased but $\mathrm{Var}[I(\omega)] \approx f(\omega)^2$ does not shrink with $n$; smoothing needed (Daniell, Welch)

**Q9.** *How does first differencing act as a filter? What does it do to the spectrum? What is the risk of overdifferencing?*

**Model answer keywords:** $(1-B)$ has transfer $|A(\omega)|^2 = 2(1-\cos 2\pi\omega)$; zero at $\omega=0$ (removes trend), gain = 4 at $\omega = 0.5$ (amplifies noise); overdifferencing injects MA(1) structure and amplifies high-frequency noise

### **Block D: Multivariate and Advanced (Questions 10–13)**

**Q10.** *What is Granger causality? Give three reasons it does not imply structural causation.*

**Model answer keywords:** past of $X$ improves linear forecast of $Y$; (1) common cause/confounding, (2) information-set dependent, (3) linear only → misses nonlinear coupling; predictive not structural

**Q11.** *Two I(1) series appear correlated with high $R^2$. How do you determine if this is spurious or cointegration?*

**Model answer keywords:** check $R^2 > DW$; run ADF on OLS residuals; if residuals are $I(0)$ → cointegrated, fit VECM; if residuals are $I(1)$ → spurious, regression is meaningless

**Q12.** *Explain GARCH(1,1). Why are GARCH returns white noise but not independent?*

**Model answer keywords:** $\sigma_t^2 = \alpha_0 + \alpha_1 X_{t-1}^2 + \beta_1\sigma_{t-1}^2$; $\mathrm{Cov}(X_t, X_{t-k}) = 0$ (uncorrelated) but $\mathrm{Cov}(X_t^2, X_{t-k}^2) \neq 0$; Ljung–Box passes on levels, fails on squares; conditional variance is time-varying

**Q13.** *Write the Kalman filter prediction and update steps. What does the Kalman gain optimize?*

**Model answer keywords:** predict $\mathbf{x}_{t|t-1} = \Phi\mathbf{x}_{t-1|t-1}$, $\mathbf{P}_{t|t-1} = \Phi\mathbf{P}\Phi' + Q$; innovation $\nu_t = y_t - A'\mathbf{x}_{t|t-1}$; gain $K = PA/S$; update $\mathbf{x}_{t|t} = \mathbf{x}_{t|t-1} + K\nu_t$; minimizes MSE; balances model prediction vs observation

### **Block E: Nonlinear and CCM (Questions 14–15)**

**Q14.** *Explain CCM. Why is the cross-map direction counterintuitive? What is the convergence criterion?*

**Model answer keywords:** if $X$ causes $Y$, then $M_Y$ contains info about $X$; cross-map from $M_Y$ to $X$; convergence = $\rho$ increases with library size $L$; counterintuitive because we map FROM the effect's manifold TO the cause

**Q15.** *Compare Granger causality and CCM across framework, assumptions, and failure modes. When should you use each?*

**Model answer keywords:** Granger = stochastic/linear/predictive; CCM = deterministic/nonlinear/dynamical; Granger fails with nonlinear coupling and synchronized systems; CCM fails with pure stochastic and very short series; use BOTH, agreement strengthens evidence
